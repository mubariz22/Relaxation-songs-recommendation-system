# Music Recommendation System for Relaxation/Meditation songs.

```python
#Importing the required libraries
import pandas as pd
```

```python
#loading datasets
tracks_data = pd.read_csv("spotify_tracks.csv")
albums_data = pd.read_csv("spotify_albums.csv")
```

```python
display(tracks_data.head())
tracks_data.columns
```

```python
display(albums_data.head())
albums_data.columns
```

```python
#drop irrelevant columns from track dataframe
tracks_df = tracks_data.drop(['Unnamed: 0','available_markets','country','disc_number','lyrics','playlist','time_signature','track_name_prev','track_number','type','artists_id','duration_ms','liveness','key'], axis=1)
tracks_df = tracks_df.rename(columns = {'id':'track_id', 'popularity': 'track_popularity'})
tracks_df.head(2)
```

```python
tracks_df['loudness'].unique()
```

```python
tracks_df.isnull().sum()
```

```python
#drop irrelevant columns from our albums dataframe
albums_df = albums_data.drop(['Unnamed: 0','album_type','artist_id','available_markets','external_urls', 'name','href','release_date_precision','total_tracks','id','track_name_prev','uri','type'], axis=1)
albums_df = albums_df.rename(columns = {'popularity': 'track_popularity'})
albums_df.head(2)
```

```python
#merging both the data frame
tracks_df = pd.merge(tracks_df, albums_df , how='inner', on = 'track_id')
```

```python
tracks_df.head()
```

```python
tracks_df.isnull().sum()
```

```python
tracks_df[tracks_df.duplicated()]
```

```python
tracks_df['meditation/relaxation music']=tracks_df['instrumentalness']>0.5
```

```python
tracks_df
```

```python
df1=tracks_df[tracks_df['meditation/relaxation music']==True]
df1
```

For meditation / relaxation music, considering

- acousticness > 0.6
- danceabbilty < 0.4
- energy < 0.4
- instrumentalness > 0.5
- liveness (Not relevant)
- speechiness < 0.33
- tempo < 125
- valence > 0.5
- key (Not relevant, mode is alread considered)
- mode= 1
- loudness <0.4
- explicit = 0

```python
#Filter the dataset based on the Audio features
df1=df1[df1['acousticness']>0.6]
df1=df1[df1['danceability']<0.4]
df1=df1[df1['energy']<0.4]
df1=df1[df1['loudness']<0.4]
df1=df1[df1['speechiness']<0.33]
df1=df1[df1['tempo']<125]
df1.head()
```

```python
#convert the release date from YYYY-MM-DD to YYYY format
df1['release_year'] = pd.to_datetime(df1.release_date).dt.year
```

```python
df1
```

```python
df1 = df1.drop(['release_date'], axis=1)
```

```python
#keep only more recent songs (songs published after 1990) to keep our dataset small.
df1[df1.release_year >= 1990]
```

```python
# preprocessing for the uri column
df1["uri"] = df1["uri"].str.replace("spotify:track:", "")
```

```python
#Reseting the index column
df1.reset_index(drop=True)
```

```python
#Save into a .csv file
df1.to_csv("track_df.csv", index=False)
```
